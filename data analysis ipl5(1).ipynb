```python
import pandas as pd
```

```python
deliveries=pd.read_csv(r"C:\Users\DELL3521\Desktop\deliveries.csv")
deliveries
```

```python
mat=pd.read_csv(r"C:\Users\DELL3521\Desktop\matches.csv")
mat
```

```python
# most number of player of matches

mat['player_of_match'].value_counts()
```

```python
# most expensive over

deliveries.groupby(['match_id','bowler','over'],as_index=False)['batsman_runs'].sum().sort_values('batsman_runs',ascending = False)
```

```python
# most teams changed by a player

mat[['id','season']].merge(deliveries,left_on = 'id',right_on = 'match_id').drop('id',axis = 1).pivot_table( index=['batsman','season','batting_team'], aggfunc=len).reset_index()[['batsman','season','batting_team']].drop_duplicates(['batsman','batting_team']).groupby('batsman')['batting_team'].count().sort_values(ascending=False)
```
