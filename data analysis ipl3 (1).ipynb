# season wise batsman runs

```python
import pandas as pd
```

```python
deliveries=pd.read_csv(r"C:\Users\DELL3521\Desktop\deliveries.csv")
deliveries
```

```python
mat=pd.read_csv(r"C:\Users\DELL3521\Desktop\matches.csv")
mat
```

```python
run=deliveries[['batsman','batsman_runs']]

runmax=run.groupby(['batsman'])[['batsman_runs']].sum().sort_values('batsman_runs',ascending=False)
runmax1=runmax[:10]

w=deliveries.rename(columns={'match_id':'id'})

t=mat[['id','season']]

season1=pd.merge(w,t,on='id')

s2=season1.pivot_table(values='batsman_runs', index='batsman', columns='season', aggfunc=sum)

s2.fillna(0,inplace=True)

s2['total']=s2[2008]+s2[2009]+s2[2010]+s2[2011]+s2[2012]+s2[2013]+s2[2014]+s2[2015]+s2[2016]+s2[2017]

s3=s2.sort_values('total',ascending=False)[:10]
s3 = s3.T

import matplotlib.pyplot as plt
% matplotlib inline

s3
```

```python
s3.iloc[:-1,:].plot(figsize=(20,7),marker='o',grid=True)
plt.title("season runs",fontsize=14)
plt.xlabel("season",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('batsman runs',fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(2008,2017)
plt.legend(fontsize ='small')
plt.show()
```
