```python
import numpy as np
```

```python
def segment(x):
    global newrfm
    if (x['rscore']in range (4,6)) & (x['fmscore'] in np.arange(4,5.5,0.5)):
        x['segmentation']='champion'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (2,6)) & (x['fmscore'] in np.arange(3,5.5,0.5)):
        x['segmentation']='loyal customer'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1) 
    if  (x['rscore']in range (3,6)) & (x['fmscore'] in np.arange(1,3.5,0.5)):
        x['segmentation']='Potential Loyalist'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (4,6)) & (x['fmscore'] in np.arange(1,1.5,0.5)):
        x['segmentation']='Recent Customers'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1) 
    if  (x['rscore']in range (3,5)) & (x['fmscore'] in np.arange(1,1.5,0.5)):
        x['segmentation']='Promising'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (2,4)) & (x['fmscore'] in np.arange(2,3.5,0.5)):
        x['segmentation']='Customers Needing Attention'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (2,4)) & (x['fmscore'] in np.arange(1,2.5,0.5)):
        x['segmentation']='About To Sleep'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (0,3)) & (x['fmscore'] in np.arange(2,5.5,0.5)):
        x['segmentation']='At Risk'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (0,2)) & (x['fmscore'] in np.arange(4,5.5,0.5)):
        x['segmentation']='Can’t Lose Them'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (1,3)) & (x['fmscore'] in np.arange(1,2.5,0.5)):
        x['segmentation']='Hibernating'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)
    if  (x['rscore']in range (0,3)) & (x['fmscore'] in np.arange(1,2.5,0.5)):
        x['segmentation']='Lost'
        df1=pd.DataFrame([[x['CustomerID'],x['recency'],x['rscore'],x['frequency'],x['fscore'],x['monetory'],x['mscore'],x['rfmscore'],x['fmscore'],x['Country'],x['segmentation']]],columns=['CustomerID','recency','rscore','frequency','fscore','monetory','mscore','rfmscore','fmscore','Country','segmentation'])
        newrfm=newrfm.append(df1)   
```
