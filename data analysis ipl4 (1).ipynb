```python
import pandas as pd
```

```python
deliveries=pd.read_csv(r"C:\Users\DELL3521\Desktop\deliveries.csv")
deliveries
```

```python
mat=pd.read_csv(r"C:\Users\DELL3521\Desktop\matches.csv")
mat
```

```python
w=deliveries.rename(columns={'match_id':'id'})

t=mat[['id','venue']]

s=pd.merge(w,t,on='id')

runs=s[s['inning']==1].groupby(['venue'])[['total_runs']].sum()

matches=mat.groupby(['venue'])[['venue']].count()

runs['match']=matches

runs['average']=runs['total_runs']/runs['match']

run=runs.iloc[[4,5,7,8,10,14,15,16,17,21,22,23,24,25,26,27,34],:]

run.sort_values('average',ascending=False,inplace=True)

run1=run[:10]

avg=pd.DataFrame(run1['average'])

def func(x):
    return float(format(x, '.2f'))

avg=avg.applymap(func)

run1['average']=avg
run1
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
ax = run1.average.plot(kind ='bar',figsize=(15,5))
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+ 1))
plt.title("venue wise average batting first",fontsize=14)
plt.xlabel("venue",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('average runs',fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(150,180)
plt.legend(fontsize ='large')
plt.show()
```
