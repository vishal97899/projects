```python
# maximum number of sixes

six=deliveries[['batsman','batsman_runs']]
w=six.loc[six['batsman_runs']==6]

sixes=w['batsman'].value_counts()[:10]
sixes=pd.DataFrame(sixes)

import matplotlib.pyplot as plt
% matplotlib inline

ax = sixes.plot(kind ='bar',figsize=(15,5),color = 'r')
for p in ax.patches:
    ax.annotate(format(p.get_height()+1), (p.get_x()+0.1, p.get_height()+ 4))
plt.title("maximum sixes",fontsize=14)
plt.xlabel("batsman",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('sixes',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='xx-large')
plt.show()
```

```python
# maximum runs in IPL

six
maxruns=six.groupby(['batsman'])[['batsman_runs']].sum().sort_values('batsman_runs',ascending=False)
maxrun=maxruns[:10]

import matplotlib.pyplot as plt
% matplotlib inline

ax = maxrun.plot(kind ='bar',figsize=(15,5),grid = True)
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+ 100))
plt.title("maximum runs in ipl",fontsize=14)
plt.xlabel("batsman",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('runs',fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(3000,5000)
plt.legend(fontsize ='xx-large')
plt.show()
```

```python
# total number of balls faced

w=six.groupby(['batsman'])[['batsman_runs']].count().sort_values('batsman_runs',ascending=False)

ballsfaced=w.rename(columns={'batsman_runs':'balls'})
ballsfaced
```

```python
df=pd.merge(maxruns,ballsfaced,on='batsman')
df
```

```python
# strikerate of top 10 batsman

e=df['batsman_runs']/df['balls']

strike=e*100

df['strike rate']=strike
df
```

```python
# maximum wickets

wicket=deliveries[['bowler','player_dismissed','dismissal_kind']]

wicket.dropna(how='any',inplace=True)
wicket
```

```python
wicket['dismissal_kind'].unique()
```

```python
# maximum wickets

hi=wicket.drop(wicket[(wicket.dismissal_kind == 'run out') | (wicket.dismissal_kind == 'retired hurt') | (wicket.dismissal_kind == 'obstructing the field')].index)

maxwickets=hi['bowler'].value_counts()
maxwickets
```

```python
maxwickets=pd.DataFrame(maxwickets)
```

```python
ax = maxwickets[:10].plot(kind ='bar',figsize=(15,5),color='g')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+ 1))
plt.title("maximum wickets in ipl",fontsize=14)
plt.xlabel("bowlers",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('wickets',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='xx-large')
plt.show()
```

```python
# maximum number of catches except the wicketkeeper

caught=deliveries[['dismissal_kind','fielder']]

caught.dropna(how='any',inplace=True)

c=caught.loc[(caught['dismissal_kind'] == 'caught')]

cb=deliveries[['bowler','dismissal_kind']]

b=cb.loc[cb['dismissal_kind'] == 'caught and bowled']

b.rename(columns={'bowler':'fielder'},inplace=True)

cab=b.groupby(['fielder'],as_index=False)[['dismissal_kind']].count()


ca=c.groupby(['fielder'],as_index=False)[['dismissal_kind']].count()

caughtw=pd.merge(ca,cab,on='fielder',how='outer')

caughtw.rename(columns={'dismissal_kind_x':'caught'},inplace=True)

caughtw.rename(columns={'dismissal_kind_y':'caught and bowled'},inplace=True)

caughtw.fillna(0,inplace=True)

caughtw['total catches']=caughtw['caught']+caughtw['caught and bowled']

caughtw.sort_values('total catches',ascending=False,inplace=True)

caughtw.drop(caughtw.index[[0,3,4,7,9,13,14,18]],inplace=True)

caughtw[:10]
```

```python
# economy of bowlers

tot=deliveries[['bowler','noball_runs','batsman_runs','bye_runs','legbye_runs','penalty_runs','extra_runs']]

tot['wide']=tot['extra_runs']-(tot['noball_runs']+tot['bye_runs']+tot['legbye_runs']+tot['penalty_runs'])

tot['runs']=tot['noball_runs']+tot['batsman_runs']+tot['wide']

totalruns=tot.groupby(['bowler'])[['runs']].sum()

count1=deliveries.loc[deliveries['ball']<=6]

count2=count1.groupby(['bowler'])[['ball']].count()

eco1=pd.merge(totalruns,count2,on='bowler')

eco1['economy']=(eco1['runs']*(6))/eco1['ball']

eco2=eco1.loc[eco1['ball']>800]

eco2['economy'].sort_values()[:10]
```

```python
# highest average batting first

avg=deliveries[deliveries['inning']==1].groupby(['batting_team'])[['total_runs']].sum().sort_values('total_runs',ascending=False)

ma=deliveries.drop_duplicates(['match_id'],keep='first')[['match_id','batting_team']]

matches=ma.groupby(['batting_team'])[['match_id']].count()

a=pd.merge(avg,matches,on='batting_team')

a.rename(columns={'match_id':'count'},inplace=True)

a['avg']=a['total_runs']/a['count']

a.sort_values('avg',ascending=False,inplace=True)

average=pd.DataFrame(a['avg'])

def func(x):
    return float(format(x, '.2f'))

average=average.applymap(func)

a['avg']=average

ax = a.avg.plot(kind ='bar',figsize=(15,5))
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+ 0.1))
a.avg.plot(kind ='bar',figsize=(15,5),colors = ['gold', '#4682B4','#FF8C00','#228B22','#4B0082','#FF0000'])
plt.title("highest average batting first",fontsize=14)
plt.xlabel("teams",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('avg runs',fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(100,200)
plt.legend(fontsize ='large')
plt.show()
```

```python
# maximum runs in death overs

dover=deliveries[['over','batsman','batsman_runs']]

dovers=dover.loc[(dover['over'])>=16]

do1=dovers.groupby(['batsman'])[['batsman_runs']].sum()

do=dovers.pivot_table(values='batsman_runs', index='batsman', columns=['over'], aggfunc=sum)

do2=pd.merge(do,do1,on='batsman')
do2.fillna(0,inplace=True)

do3=do2.sort_values('batsman_runs',ascending=False)[:10]

do3.iloc[:,:-1].plot(kind ='bar',figsize=(20,7),colors = ['gold', '#4682B4','#FF8C00','#228B22','#4B0082','#FF0000'])
plt.title("death overs",fontsize=14)
plt.xlabel("batsman",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('batsman_runs',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='xx-large')
plt.show()
```

```python
do4=do3.T
```

```python
do4.iloc[:-1,:].plot(figsize=(15,7),color = ['gold', '#4682B4','#FF8C00','#228B22','#4B0082','#FF0000'],marker='o',grid=True)
plt.title("deathover runs",fontsize=14)
plt.xlabel("overs",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('runs',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='large')
plt.show()
```

```python
# maximum runs in powerplay overs

max1=deliveries[deliveries['over']<=6].pivot_table(index='batsman',columns='over',values='batsman_runs',aggfunc='sum')

max1.fillna(0,inplace=True)

max1['total']=max1[1]+max1[2]+max1[3]+max1[4]+max1[5]+max1[6]

max1.sort_values('total',ascending=False,inplace=True)
max2=max1[:10]

import matplotlib.pyplot as plt
% matplotlib inline

max2.iloc[:,:-1].plot(kind ='bar',figsize=(15,7),color = ['gold', '#4682B4','#FF8C00','#228B22','#4B0082','#FF0000'])
plt.title("powerplay Runs",fontsize=14)
plt.xlabel("batsman",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('batsman Runs',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='large')
plt.show()
```

```python
# maximum wickets in powerplay overs

wic=deliveries[deliveries['over']<=6][['over','bowler','dismissal_kind']]

wic1=wic.loc[~wic['dismissal_kind'].isin(['retired hurt','obstructing the field','run out'])]

wic2=wic1.dropna(how='any')

wic3=wic2.pivot_table(index='bowler',values='dismissal_kind',columns='over',aggfunc=len)

wic4=wic3.fillna(0)

wic4['total']=wic4[1]+wic4[2]+wic4[3]+wic4[4]+wic4[5]+wic4[6]

wic4.sort_values('total',ascending=False,inplace=True)
wic5=wic4[:10]

import matplotlib.pyplot as plt
% matplotlib inline

wic5.iloc[:,:-1].plot(kind ='bar',figsize=(15,7),color = ['gold', '#4682B4','#FF8C00','#228B22','#4B0082','#FF0000'])
plt.title("powerplay wickets",fontsize=14)
plt.xlabel("bowler",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('wickets',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='large')
plt.show()
```

```python
# maximum wickets in death overs

wi=deliveries[deliveries['over']>=16][['over','bowler','dismissal_kind']]

wi1=wi.loc[~wi['dismissal_kind'].isin(['retired hurt','obstructing the field','run out'])]

wi2=wi1.dropna(how='any')

wi3=wi2.pivot_table(index='bowler',values='dismissal_kind',columns='over',aggfunc=len)

wi3.fillna(0,inplace=True)

wi3['total']=wi3[16]+wi3[17]+wi3[18]+wi3[19]+wi3[20]

wi3.sort_values('total',ascending=False,inplace=True)
wi4=wi3[:10]

wi5=wi4.T

import matplotlib.pyplot as plt
% matplotlib inline

wi4.iloc[:,:-1].plot(kind ='bar',figsize=(15,7),color = ['gold', '#4682B4','#FF8C00','#228B22','#4B0082','#FF0000'])
plt.title("deathover wickets",fontsize=14)
plt.xlabel("bowler",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('wickets',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='large')
plt.show()
```

```python
wi5.iloc[:-1,:].plot(figsize=(15,7),color = ['gold', '#4682B4','#FF8C00','#228B22','#4B0082','#FF0000'],marker='o',grid=True)
plt.title("deathover wickets",fontsize=14)
plt.xlabel("overs",fontsize=14)
plt.xticks(fontsize=14, rotation=90)
plt.ylabel('wickets',fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize ='large')
plt.show()
```
